Task-2

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load the Sales dataset
sales_df = pd.read_csv('sales.csv')

# Add the 'No_product' column to the dataset
sales_df['No_product'] = [random.randint(1, 10) for _ in range(len(sales_df))]

# Plot a bar graph for Unit_Cost
plt.figure(figsize=(10, 6))
plt.bar(sales_df.index, sales_df['Unit_Cost'])
plt.xlabel('Index')
plt.ylabel('Unit Cost')
plt.title('Unit Cost Distribution')
plt.show()

# Plot a bar graph for No_product
plt.figure(figsize=(10, 6))
plt.bar(sales_df.index, sales_df['No_product'])
plt.xlabel('Index')
plt.ylabel('Number of Products')
plt.title('Number of Products Distribution')
plt.show()
```

Task-3

```python
# Plot a pie chart for Item_Type
item_type_counts = sales_df['Item_Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(item_type_counts, labels=item_type_counts.index, autopct='%1.1f%%')
plt.title('Item Type Distribution')
plt.show()

# Plot a pie chart for No_product
no_product_counts = sales_df['No_product'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(no_product_counts, labels=no_product_counts.index, autopct='%1.1f%%')
plt.title('Number of Products Distribution')
plt.show()
```

Task-4

```python
# Plot a pie chart for Region
region_counts = sales_df['Region'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%')
plt.title('Region Distribution')
plt.show()

# Add random values to the Gross_margin column
sales_df['Gross_margin'] = [random.uniform(0.0, 0.5) for _ in range(len(sales_df))]

# Plot a pie chart for Gross_margin
plt.figure(figsize=(8, 8))
plt.pie(sales_df['Gross_margin'], labels=sales_df.index, autopct='%1.1f%%')
plt.title('Gross Margin Distribution')
plt.show()
```

```python
# Task 3: Plot a pie chart for Item_Type and No_product columns
item_type_counts = sales_df['Item_Type'].value_counts()
no_product_counts = sales_df['No_product'].value_counts()

# Create a subplot for Task 3 and Task 4
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Task 3: Item_Type and No_product
axes[0].pie(item_type_counts, labels=item_type_counts.index, autopct='%1.1f%%')
axes[0].set_title('Item Type Distribution')

axes[1].pie(no_product_counts, labels=no_product_counts.index, autopct='%1.1f%%')
axes[1].set_title('Number of Products Distribution')

# Task 4: Region and Gross_margin
region_counts = sales_df['Region'].value_counts()
gross_margin_counts = sales_df['Gross_margin'].value_counts()

# Create a subplot for Task 3 and Task 4
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Task 4: Region and Gross_margin
axes[0].pie(region_counts, labels=region_counts.index, autopct='%1.1f%%')
axes[0].set_title('Region Distribution')

axes[1].pie(gross_margin_counts, labels=gross_margin_counts.index, autopct='%1.1f%%')
axes[1].set_title('Gross Margin Distribution')

plt.show()
```
